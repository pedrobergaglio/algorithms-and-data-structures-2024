    aux eltosRangoRecursivo(in n:nodo, in desde : Z, in hasta : Z){

        int res = 0
        if n.valor = null 
            return 0
        else if desde <= n.valor < hasta res++

        if n.izq != null and n.valor > desde
            res += eltosRangoRecursivo(n.izq, desde, hasta)
        if n.der != null and n.valor < hasta
            res += eltosRangoRecursivo(n.der, desde, hasta)
        
        return res

    }

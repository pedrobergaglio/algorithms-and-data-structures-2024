### Ejercicio 1

Nodo<T> es Struct<dato: T, izq: Nodo, der: Nodo >

Modulo ArbolBinario<T> implementa Arbol Binario<T>{

    var raiz: Nodo<T> // “puntero” a la raiz del arbol

    pred InvRep natural (c:AB<T>){
        1. para cada nodo, si no tiene valor, no tiene hijos, recursivo
        2. no hay repetidos, recursivo
        //esto no: 3. para cada nodo, su hijo izquierdo es mas chico, su hijo derecho es mas grande, recursivo
        4. para cada nodo, su cantidad es igual a 1 + la cantidad de sus hijos, recursivo
    }

    pred InvRep (c:AB<T>){

        c.raiz.conHijosNoNulo() and
        opcion 1 (descartada):
            forall e:T :: c.pertenece(e) ==>L !c.eliminar(e).pertenece(e) 
            //suponiendo que eliminar solo eliminar un ejemplar
        opcion 2:
            forall e, f : T :: (c.pertenece(e) and c.pertenece(f)) ==>L e!=f

        and
        c.raiz.cantidadActualizada()

    }

    pred cantidadActualizada(n: Nodo) {
        n == null or
        (
            n.cant = 1 +
                ifthenelse (n.izq = null, 0, n.izq.cant) +
                ifthenelse (n.der = null, 0, n.der.cant)
            and
            cantActualizado(n.izq) 
            and
            cantActualizado(n.der) 
        )
    }

    pred conHijosNoNulo(n: nodo):bool{
        if (n == null && (n.izq != null || n.der != null)): 
            return false
        else if (c.raiz.izq != null && c.raiz.der != null):
            return n.izq.conHijosNoNulo() && n.der.conHijosNoNulo()
        else if (c.raiz.izq != null):
            return n.izq.conHijosNoNulo()
        else if (c.raiz.der != null):
            return n.der.conHijosNoNulo()
        else:
            return true
    }



    pred Abs (in c: AB<T>, out c': TAD<T>){
    1. c'.raiz es la misma que c.raiz  // La raíz en la implementación representa la raíz en el TAD.
    2. para todo nodo `n` en el árbol `c`, `n` también existe en `c'`
    3. para cada nodo `n` de `c`, sus hijos `izq` y `der` en la implementación corresponden a sus hijos en el TAD
    4. el tamaño del árbol `c` en la implementación coincide con el tamaño abstracto de `c'`
    }


    proc altura(in ab: ArbolBinario<T>): int {
        // devuelve la distancia entre la ra´ız y la hoja m´as lejana

        return alturaRecursivo(ab.raiz)
    
    }

- complejidad: O(n) con n cantidad de nodos

    proc alturarecursivo (n:nodo): int{

        if ab.raiz == null: return 0

        int suma = 0
        int izq = alturarecursivo(ab.raiz.izq)
        int der = alturarecursivo(ab.raiz.der)

        if izq < der
            suma = der
        else: suma = izq

        return 1 + suma
    }

    proc cantidadHojas(in ab: ArbolBinario<T>): int {

        return cantidadHojasRecursivo(ab.raiz)

    }

- complejidad: O(n) con n cantidad de nodos

    proc cantidadHojasRecursivo(n:nodo):int{
        
        if n == null return 0
        if n.izq == null and n.der == null return 1

        return cantidadHojasRecursivo(n.izq) + cantidadHojasRecursivo(n.der)
    }

    proc esta(in ab: ArbolBinario<T>, int t: T): bool {
        
        // devuelve true si el elemento esta en el arbol

        return estaRecursivo(ab.raiz, t)

    }

- complejidad: O(n) con n cantidad de nodos

    proc estaRecursivo(n:nodo, t:T):bool{

        if n == null return false
        if n == t return true

        return estaRecursivo(n.izq, t) or estaRecursivo(n.der, t)

    }

    proc cantidadApariciones(in ab: ArbolBinario<T>, int t: T): int {

        return CantRecur(ab.nodo, t)
        
    }

- complejidad: O(n) con n cantidad de nodos

    proc CantRecur(n:nodo, t:T): int {

        if n == null return 0

        if n.valor == t
            return 1 + CantRecur(n.izq, t) + CantRecur(n.der, t)
        else:
            return CantRecur(n.izq, t) + CantRecur(n.der, t)

    }

}

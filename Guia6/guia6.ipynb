- complejidad: O( c.datos.len * 2)

proc agregar (inout c:conjArr, in e:T){

    if c.datos.len == tamaño: return
    if c.pertenece(e): return

    Array nuevo = new Array[c.datos.len +1]
    for i<c.datos.len
        nuevo[i] = c.datos[i]
    nuevo[-1] = e
    c.datos = nuevo

}

- complejidad: O( c1.datos.len * c2.datos.len)

proc intersecar (inout c1: ConjArr, c2: ConjArr){

    Array nuevo = new Array[0]<T>

    for i < c1.datos.len
        if c2.pertenece(c1[i]):
            nuevo.agregar(c1[i])
    
    c1.datos = nuevo
}

- complejidad: O( c.datos.len * 2)

proc sacar (inout c:conjArr, in e:T){ 

    if !c.pertenece(e): return

    Array<T> nuevo = new Array<T>[c.datos.len-1]

    int k = 0

    for i < c.datos.len
        if c.datos[i] != e
            nuevo[k] = c.datos[i]
            k++
    
    c.datos = nuevo

}

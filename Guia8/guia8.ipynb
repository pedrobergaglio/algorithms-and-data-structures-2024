Modulo Matriz implementa Matriz infinita de booleanos

    var filas vector<vector<boolean>>
    var complementado boolean

    proc crear O(1)
        var s = new Matriz()
        s.complementado = false
        s.filas = new vectorVacio()
        return s

    proc asignar(m, f, c, v:boolean) // O(f(n)*(f + c))

        while f > m.filas.longitud + 1
            var columna = new vectorVacio() //O(1)
            m.filas.agregarAtras(columna) //O(f(n))

        while c > m.filas[f].longitud + 1
            m.filas[f].agregarAtras(false) //O(f(n))
        
        m.filas[f].modificarPosicion(v)


    proc complementar O(1)
        complementado = !complementado

    proc ver(m, f, c) O(1)

        if f > m.filas.longitud-1 || c > m.filas[f].longitud - 1
            return complementado

        else if complementado
            return !m[f][c]
        else
            return m[f][c]
